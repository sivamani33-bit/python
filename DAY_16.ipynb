```python
import seaborn as sns
import numpy as np
from scipy.stats import zscore

data=sns.load_dataset('iris')
df=data.copy()

z_scores = np.abs(zscore(df.drop('species',axis=1)))
```

```python
z_scores
```

```python
zscore(df.drop('species',axis=1))
```

```python
thresold = 3
non_outliers = (z_scores < thresold).all(axis=1)
df_no_outliers = df[non_outliers]
```

```python
df_no_outliers
```

```python
thresold = 3
outliers = (z_scores > thresold)
outlier = df[outliers.any(axis=1)]
outlier
```
